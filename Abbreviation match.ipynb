# Find abbreviations used

## Imports and functions

```python
import pandas as pd
from itertools import combinations
import difflib
import jupyter_black

jupyter_black.load()


def Sorting(lst):
    """
    sort lists based on length
    Args:
        lst: list of strings
    Returns:
        list
    """
    lst = sorted(lst, key=len)
    return lst
```

## Read files, setup variables

```python
label_choice = []
target_label_choice = []
name_set = []
original_index = []
original_label = []

k = 0

df = pd.read_csv("sample_project.csv")
```

## Run function

```python
for index, row in df.iterrows():
    nacount = 0
    name_par = row["name"].split("_")
    label_par = row["label"].replace(" / ", " ").split(" ")
    # all posible label combination
    sample_list = label_par
    list_combinations = list()

    for n in range(len(sample_list) + 1):
        list_combinations += list(combinations(sample_list, n))
    list_strings = []
    for i in list_combinations:
        list_strings.append(" ".join(i))
    # find the close match

    for m in name_par:
        # loose match
        label_choice_list = difflib.get_close_matches(m, list_strings, n=4, cutoff=0)
        sort_label = Sorting(label_choice_list)
        label_choice.append(sort_label)

        # close match
        for x in sort_label:
            x_splited = list(x.lower().replace(" ", ""))
            index = -1
            count = 0
            for l in list(m.lower()):
                if l in x_splited[index + 1 :]:
                    index = x_splited.index(l)
                    count += 1
                else:
                    nacount += 1
                    break

            if nacount == len(sort_label):
                target_label_choice.append("NA")

            if count == len(m):
                target_label_choice.append(x)
                break
            else:
                continue

        name_set.append(m)
        original_index.append(k)
        original_label.append(row["label"])

    k += 1
```

## Output

```python
column_lsit = {
    "label_choice": label_choice,
    "target_label_choice": target_label_choice,
    "name_set": name_set,
    "original_index": original_index,
    "original_label": original_label,
}
result = pd.DataFrame(data=column_lsit)
result
```
